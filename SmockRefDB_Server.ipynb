```python
seq_dict = {}
count = 0
for row in seq_names_list:
            count += 1
            taxid = row[0].split(sep = "|")
            seq_dict.setdefault(taxid[1],[]).append(int(row[1]))
print(count)
print(len(list(seq_dict.keys())))
```

```python
with open(data_root + 'Describe_DB_filterd.txt', 'w') as thefile:
    c1 = 0
    c2 = 0
    keys = list(seq_dict.keys()) 
    for n, key in enumerate(keys):
        c1 +=1        
        sub_reds_lens = []
        for i in range(len(seq_dict[key])):
            sub_reds_lens.append(seq_dict[key][i])  
        if len(sub_reds_lens) == 1:
            continue
        thefile.write("Key : |{0}|\n".format(key))
        thefile.write("Number of reads: {0}\n".format(len(sub_reds_lens)))
        thefile.write("Max len: {0}\n".format(max(sub_reds_lens)))
        thefile.write("Min len: {0}\n".format(min(sub_reds_lens)))
        thefile.write("-"*50 + "\n")
        c2 += 1
    print("{0}- items iterated".format(c1))
    print("{0}- items described".format(c2))
```
